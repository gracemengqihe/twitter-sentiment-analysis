```
'''Split training and testing data set'''
X_train = [training_data[i][1] for i in range(len(training_data))]
y_train = [training_data[i][0] for i in range(len(training_data))]
X_test = [testing_data[i][1] for i in range(len(testing_data))]
y_test = [testing_data[i][0] for i in range(len(testing_data))]
```

```
'''Remove tags
Eliminate unwanted characters, such as HTML tags, punctuation marks, special characters, white spaces etc.
The username in the format of '@user' should be removed since it doesn't have much impact on the final outcome
The HTML tags should also be removed since it doesn't have much impact on the final outcome
'''

import re
def removeTags(tweet):

    # Remove urls
    tweet = re.sub(r"http\S+|www\S+|https\S+", '', tweet, flags=re.MULTILINE)
    # ASCII Arrows
    tweet = re.sub(r"""[\-]+>|<[\-]+""",'', tweet)
    # Twitter username:
    tweet = re.sub(r"""(?:@[\w_]+)""",'', tweet)
    # Twitter hashtags:
    tweet = re.sub(r"""(?:\#+[\w_]+[\w\'_\-]*[\w_]+)""",'', tweet)
    # email addresses
    tweet = re.sub(r"""[\w.+-]+@[\w-]+\.(?:[\w-]\.?)+[\w-]""",'', tweet)

    return tweet

X_train = [removeTags(s) for s in X_train]
y_train = [removeTags(s) for s in y_train]
X_test = [removeTags(s) for s in X_test]
y_test = [removeTags(s) for s in y_test]

print("------------------------------------")
print("Sample Data")
print("After removing tags: {}".format(X_train[1]))
print("------------------------------------")
```

```
'''
function to convert numbers to words
Return : text which have all numbers or integers in the form of words
Input : string
Output : string
'''
import num2words 
def num_to_words(text):
  
  no_num_text = []
	# splitting text into words with space
  for sent in text:
    text_split = sent.split()

    for i in range(len(text_split)):
      if text_split[i].isdigit():
        # use attribute num2words function to return the number as a word
        text_split[i] = num2words.num2words(text_split[i])

      # joining list into string with space
    numbers_to_words = ' '.join(text_split)
    no_num_text.append(numbers_to_words)
  return no_num_text


X_train = num_to_words(X_train)
y_train = num_to_words(y_train)
X_test = num_to_words(X_test)
y_test = num_to_words(y_test)

print("------------------------------------")
print("Sample Data")
print("After converting numbers to words: {}".format(X_train[1]))
print("------------------------------------")
```

```
'''
Remove anything that is not a word(letter a-z & A-Z) by using regular expression
'/s' is used to match any Unicode whitespace characters in tweets such as [\t\n\r\f\v] 
'''
def remove_punctuation_re(text):
    # remove non-words items 
    new_string = re.sub('[^a-zA-Z]',' ',str(text))
    # remove unicode whitespace characters
    no_punct = re.sub('\s+',' ', new_string)
    return no_punct

X_train = [remove_punctuation_re(s).strip() for s in X_train] # strip() is used to remove the extra white spaces in the beginning and end of the string
y_train = [remove_punctuation_re(s).strip() for s in y_train]
X_test = [remove_punctuation_re(s).strip() for s in X_test]
y_test = [remove_punctuation_re(s).strip() for s in y_test]

print("------------------------------------")
print("Sample Data")
print("After removing anything that is not a word: {}".format(X_train[1]))
print("------------------------------------")
```

```
'''Encode Twitter data '''
# Define the sequence length 
seq_length = 9

def encode_and_add_padding(sentences, seq_length, word_index):
    sent_encoded = []
    for sent in sentences:
      temp_encoded = []
      for word in sent:

        #if the word is in the word_index dictionary
        if word in word_index:
          temp_encoded.append(word_index[word])
        #if the word is not in the word_index dictionary 
        else: 
          temp_encoded.append(word_index['[UNKNOWN]'])

      # check the temp_encode length 
      # if the length is less than the seq_length, extra pad is added to the end  
      if len(temp_encoded) < seq_length:
        temp_encoded += [word_index['[PAD]']] * (seq_length - len(temp_encoded))
      
      # if the length is longer than the seq_length, the encoded list is chopped 
      if len(temp_encoded) >= seq_length:
        temp_encoded = temp_encoded[:seq_length]       
      sent_encoded.append(temp_encoded)
    return sent_encoded

X_train_pad_encoded = encode_and_add_padding(X_train_le, seq_length, word_index)
X_test_pad_encoded = encode_and_add_padding(X_test_le, seq_length, word_index)
```

### 2.1.2. Build Word Embeddings Model

*You are required to describe how hyperparameters were decided with justification of your decision.*

```
'''make window size = 2 for skip-gram
Since the vector size length is 9, window size of 2 is able to capture the context of the entire sentence 
'''

skip_grams = []

for sent in X_encoded: 
    for i in range(len(sent)):
        centre = sent[i]
        # when there is at least 2 words before and after the centre word
        if i > 1 and i < len(sent) - 2:
          context = sent[i-2:i] + sent[i+1:i+3]
        # when there is less than or 2 words before the centre word 
        elif i <= 1: 
          context = sent[0:i] + sent[i+1:i+3]
        # when there is less than 2 words after the centre word 
        elif i >= len(sent) - 2:
          context = sent[i-2:i] + sent[i+1:] 
        # split the context word to match with centre word. Expected skipgram format: (centre, context[0]), (centre, context[1])..
        for w in context:
          skip_grams.append([centre, w])
```

```
### SkipGram 1###
'''hyperparameter
voc_size: Vocabulary size of the training data.
embedding_size: Word embedding size.
learning_rate: Rate for weight update equations
'''
voc_size = len(word_list)
learning_rate = 0.01
batch_size = 10000
embedding_size = 20
no_of_epochs = 50
```

```
### SkipGram 1###
'''prepare batch from skip-gram'''
voc_size = len(word_list)

def prepare_batch(data_temp):
    inputs = []
    labels = []

    for i in range(len(data_temp)):
        input_temp = [0]*voc_size
        # centre word (input)
        input_temp[data_temp[i][0]] = 1 
        inputs.append(input_temp)
        # context words (prediction)
        labels.append(data_temp[i][1])  

    return np.array(inputs), np.array(labels)
```

```
### SkipGram 1###
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from random import shuffle
from sklearn.metrics import accuracy_score
import matplotlib
import matplotlib.pyplot as plt

class SkipGram(nn.Module):

    def __init__(self):
        super(SkipGram, self).__init__()

        self.linear1 = nn.Linear(voc_size, embedding_size, bias=False)
        self.linear2 = nn.Linear(embedding_size, voc_size, bias=False)

        self.linear1.weight.data.uniform_(-1,1)
        self.linear2.weight.data.uniform_(-1,1)

    def forward(self, x):
        hidden = self.linear1(x)
        out = self.linear2(hidden)
        return out

skipgram_model = SkipGram()
criterion = nn.CrossEntropyLoss()
optimiser = optim.SGD(skipgram_model.parameters(), lr=learning_rate)

for epoch in range(no_of_epochs):
  
  for ind in range(0,len(skip_grams),batch_size):
    data_temp = skip_grams[ind : min(ind+batch_size, len(skip_grams))]
    inputs_temp, labels_temp = prepare_batch(data_temp)

    inputs_torch = torch.from_numpy(inputs_temp).float()
    labels_torch = torch.from_numpy(labels_temp)

    optimiser.zero_grad()    
    outputs = skipgram_model(inputs_torch)

    loss = criterion(outputs, labels_torch)
    
    loss.backward()
    optimiser.step()
    
  print('Epoch: %d, loss: %.4f' %(epoch + 1, loss))

weight1 = skipgram_model.linear1.weight

# Get the trained projection matrix
trained_embeddings = weight1.detach().T.numpy()
```

```
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from sklearn.metrics import accuracy_score

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class Bi_LSTM_Emb(nn.Module):
    def __init__(self):
        super(Bi_LSTM_Emb, self).__init__()
        self.emb = nn.Embedding(vocab_size, concat_matrix.shape[1])
        # Initialize the Embedding layer with the concatenated embedding matrix 
        self.emb.weight.data.copy_(concat_matrix)
        # Optional: set requires_grad = False to make this lookup table untrainable
        self.emb.weight.requires_grad = False

        self.lstm = nn.LSTM(concat_matrix.shape[1], n_hidden, batch_first =True, bidirectional=True)
        self.linear = nn.Linear(n_hidden*2, n_class)

    def forward(self, x):
        # Get the embeded tensor
        x = self.emb(x)        
        # we will use the returned h_n of shape (num_layers * num_directions, batch, hidden_size): tensor containing the hidden state for t = seq_len.
        # details of the outputs from nn.LSTM can be found from: https://pytorch.org/docs/stable/generated/torch.nn.LSTM.html
        lstm_out, (h_n,c_n) = self.lstm(x)
        # concat the last hidden state from two direction
        hidden_out =torch.cat((h_n[0,:,:],h_n[1,:,:]),1)
        z = self.linear(hidden_out)
        return z
```

```
'''Word analogy task 
when A, B, C are given, we aim to find D 
A is to B what C is to D.
(e.g. man is to king what women is to D.)
'''
def evaluate_vectors(W, vocab):
    """Evaluate the trained word vectors on a variety of tasks
    The semantic questions include analogies about people or places  
    The syntactic questions include analogies about verb tenses or forms of adjectives  
    """

    file_names = [
        'capital-common-countries.txt', 'capital-world.txt', 'currency.txt',
        'city-in-state.txt', 'family.txt', 'gram1-adjective-to-adverb.txt',
        'gram2-opposite.txt', 'gram3-comparative.txt', 'gram4-superlative.txt',
        'gram5-present-participle.txt', 'gram6-nationality-adjective.txt',
        'gram7-past-tense.txt', 'gram8-plural.txt', 'gram9-plural-verbs.txt',
        ]

    # to avoid memory overflow, could be increased/decreased
    # depending on system and vocab size
    split_size = 10

    semantic_correct = 0; # count correct semantic questions
    syntactic_correct = 0; # count correct syntactic questions
    total_correct = 0 # count correct questions
    semantic_count = 0; # count all semantic questions
    syntactic_count = 0; # count all syntactic questions
    total_count = 0 # count all questions
    full_count = 0 # count all questions, including those with unknown words

    for i in range(len(filenames)):
      with open(filenames[i], 'r') as f:
        # Strip whitespace characters from the end of the string and then split by whitespace 
        full_data = [line.rstrip().split(' ') for line in f]
        # Check the total numbers of questions 
        full_count += len(full_data)
        # Store the question into the data list if all words in the question are in the word_index 
        data = [x for x in full_data if all(word in vocab for word in x)]

      # If there is no question have a complete set of words been covered in the word_index
      if len(data) == 0:
        print("ERROR: no lines of vocab kept for %s !" % filenames[i])
        print("Example missing line:", full_data[0])
        continue
      
      indices = np.array([[int(vocab[word]) for word in row] for row in data])
      ind1, ind2, ind3, ind4 = indices.T
      
      predictions = np.zeros((len(indices),))
      num_iter = int(np.ceil(len(indices) / float(split_size)))
     
      for j in range(num_iter):
          subset = np.arange(j*split_size, min((j + 1)*split_size, len(ind1)))

          pred_vec = (W[ind2[subset], :] - W[ind1[subset], :] + W[ind3[subset], :])

          #cosine similarity if input W has been normalized
          dist = np.dot(W, pred_vec.T)

          for k in range(len(subset)):
              dist[ind1[subset[k]], k] = -np.Inf
              dist[ind2[subset[k]], k] = -np.Inf
              dist[ind3[subset[k]], k] = -np.Inf

          # predicted word index
          predictions[subset] = np.argmax(dist, 0).flatten()
      
      val = (ind4 == predictions) # correct predictions

      total_count = total_count + len(ind1)
      total_correct = total_correct + sum(val)
      if i < 5:
          semantic_count = semantic_count + len(ind1)
          semantic_correct = semantic_correct + sum(val)
      else:
          syntactic_count = syntactic_count + len(ind1)
          syntactic_correct = syntactic_correct + sum(val)

      print("%s:" % filenames[i])
      print('ACCURACY TOP1: %.2f%% (%d/%d)' %
          (np.mean(val) * 100, np.sum(val), len(val)))
      
    return semantic_correct, syntactic_correct, total_correct, semantic_count, syntactic_count, total_count, full_count
```

```

correct_sem, correct_syn, correct_tot, count_sem, count_syn, count_tot, full_count = evaluate_vectors(shared_trained_embeddings, word_index)
print('Questions seen/total: %.2f%% (%d/%d)' %
    (100 * count_tot / float(full_count), count_tot, full_count))
print('Semantic accuracy: %.2f%%  (%i/%i)' %
    (100 * correct_sem / float(count_sem), correct_sem, count_sem))
print('Syntactic accuracy: %.2f%%  (%i/%i)' %
    (100 * correct_syn / float(count_syn), correct_syn, count_syn))
print('Total accuracy: %.2f%%  (%i/%i)' % (100 * correct_tot / float(count_tot), correct_tot, count_tot))
```
